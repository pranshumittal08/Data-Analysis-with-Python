# Bollywood Data Set Practice

```python
#import major libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv("Bollywood Movies.csv", index_col = "S No")
movies_df = pd.DataFrame(data)
pd.set_option("display.max_rows", 150)
pd.set_option("display.max_columns", 15)
movies_df.head(100)
```

```python
movies_df.shape
```

```python
movies_df.tail()
```

```python
movies_df.dropna(axis = 0, inplace = True)
```

```python
movies_df.shape
```

```python
movies_df.ndim
```

```python
movies_df.size
```

```python
movies_df.dtypes
```

```python
#Converting the index type of the dataframe from float to int
movies_df.index = movies_df.index.astype(np.int16)
```

```python
#Reducing the memory space occupied by youtube views, likes and comments columns
movies_df[["Youtube Views","Youtube Likes","Youtube Dislikes"]] = movies_df[["Youtube Views","Youtube Likes","Youtube Dislikes"]].astype(np.int32)
```

```python
movies_df.memory_usage()
```

#### Question 2

```python
movies_df['Genre - Defined'].value_counts()
```

```python
movies_df["Genre - Defined"] = movies_df["Genre - Defined"].str.strip()
movies_df["Genre - Defined"].unique()
```

```python
movies_df['Genre - Defined'].value_counts()
```

```python
#Plotting a bar and point plot to show the number of releases in each Genre
sn.pointplot(movies_df['Genre - Defined'].value_counts().index, y = movies_df['Genre - Defined'].value_counts(),linestyles='dotted', color = 'blue')
sn.barplot(movies_df['Genre - Defined'].value_counts().index, y = movies_df['Genre - Defined'].value_counts())
plt.title("Genre wise Movie Counts", fontsize = 16)
plt.xlabel("Genres", fontsize = 12)
plt.ylabel("Counts", fontsize = 12)
plt.show()
```

### From the above plots it is evident that Maximum movies are released in Comedy Genre and Minimum in Action Genre

#### Question 3

```python
pd.crosstab(index = movies_df["Genre - Defined"], columns = movies_df["Release Date (N / LW / Festive)"])
```

```python
pd.crosstab(index = movies_df["Genre - Defined"], columns = movies_df["Release Date (N / LW / Festive)"], values = movies_df["Budget"], aggfunc = np.median).round(2)
```

```python
pd.crosstab(index = movies_df["Genre - Defined"], columns = movies_df["Release Date (N / LW / Festive)"], 
            values = movies_df["Box Office Collection"], aggfunc = np.average).round(2)
```

It is observed from the above table that-
    1. Action Movies earn more during the festive seasons in comparison to normal days
    2. Comedy movies perform better when released on Long Weekend
    3. Romantic Movies do not perform well durig festive season

## Questions 13-20

A retrospective sample of males in a heart-disease high-risk region
of the Western Cape, South Africa. There are roughly two controls per
case of CHD. Many of the CHD positive men have undergone blood
pressure reduction treatment and other programs to reduce their risk
factors after their CHD event. In some cases the measurements were
made after these treatments. These data are taken from a larger
dataset, described in  Rousseauw et al, 1983, South African Medical
Journal. 

sbp		systolic blood pressure
tobacco		cumulative tobacco (kg)
ldl		low densiity lipoprotein cholesterol
adiposity
famhist		family history of heart disease (Present, Absent)
typea		type-A behavior
obesity
alcohol		current alcohol consumption
age		age at onset
chd		response, coronary heart disease


```python
saheart = pd.read_csv("SAheart.data",index_col = 'row.names' )
sa_df = pd.DataFrame(saheart)
pd.set_option("display.max_rows",500)
sa_df.head(462)
```

```python
sa_df[["age", "sbp"]].corr()
```

```python
sn.regplot(sa_df["age"],sa_df["sbp"])
plt.show()
```

```python
sn.jointplot(sa_df["age"],sa_df["sbp"], kind = 'reg')
plt.show()
```

```python
sn.distplot(a= sa_df[sa_df["chd"]==1]["tobacco"], kde = True, label = "Have CHD")
sn.distplot(a= sa_df[sa_df["chd"]==0]["tobacco"], kde = True, label = "Does not have CHD")
plt.xticks(np.arange(0,21,2))
plt.xlim(-4,20)
plt.legend()
plt.show()
```

#### 1. More than 40% of the people who do not have CHD consumed almost no tobacco in their liftime
#### 2. As the distribution of people who have CHD and consume tobacco is positive skewed it can assesed that people who have CHD consumed more than 2Kg of tobacco till date
